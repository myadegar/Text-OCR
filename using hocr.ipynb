```python
import pandas as pd
import math
from operator import itemgetter

def hocr_to_dataframe(fp):

    from lxml import etree
    import pandas as pd
    import os

    doc = etree.parse(fp)
    words = []
    wordConf = []
    X1 = []
    Y1 = []
    X2 = []
    Y2 = []

    for path in doc.xpath('//*'):
        if 'ocrx_word' in path.values():
            conf = [x for x in path.values() if 'x_wconf' in x][0]
            wordConf.append(int(conf.split('x_wconf ')[1]))
            
            bbox = [x for x in path.values() if 'bbox' in x][0]
            X1.append(int(bbox.split('bbox ')[1].split(';')[0].split(' ')[0]))
            Y1.append(int(bbox.split('bbox ')[1].split(';')[0].split(' ')[1]))
            X2.append(int(bbox.split('bbox ')[1].split(';')[0].split(' ')[2]))
            Y2.append(int(bbox.split('bbox ')[1].split(';')[0].split(' ')[3]))
            
            words.append(path.text)

    dfReturn = pd.DataFrame({'word' : words,
                             'confidence' : wordConf,
                             'X1' : X1,
                             'Y1' : Y1,
                             'X2' : X2,
                             'Y2' : Y2})

    return(dfReturn)


```

```python
dataframe = hocr_to_dataframe('./Untitled Folder/Befundschein 2019-10-25_page-001.hocr')
dataframe.to_excel("./Untitled Folder/Befundschein 2019-10-25_page-001.xlsx")
```

```python
excel_file = "./Untitled Folder/Befundschein 2019-10-25_page-001.xlsx"
datas = pd.read_excel(excel_file)

X1_list = list(datas['X1'])
Y1_list = list(datas['Y1'])

first_line = min(Y1_list)
last_line = max(Y1_list)

first_col = min(X1_list)

```

```python
counter = 0
tole = 2
y1 = first_line
lines = []
while y1 <= last_line:
    
    aa = datas.loc[(datas['Y1'] >= first_line + 40*counter - tole)]
    aa = aa.loc[(aa['Y1'] < first_line + 40*(counter+1) + tole), ['word', 'X1', 'X2']]
    
    lines.append(aa.values.tolist())
    
    y1 = first_line+ (40*(counter+1))
    counter += 1
```

```python
with open('./Untitled Folder/Befundschein 2019-10-25_page-001.txt', 'w', encoding="utf-8") as text:
    
    for i in range(len(lines)):

        for j in range(len(lines[i])):
            
            # sort list according to X1 increasingly
            line = sorted(lines[i], key=itemgetter(1)) 

            if j == 0:

                space = math.ceil((line[j][1])/13)
                for k in range(space):
                    text.write(' ')
                text.write(line[j][0])

            elif j > 0:

                space = math.ceil((line[j][1]-line[j-1][2])/13)
                #print(space)
                for k in range(space):
                    text.write(' ')

                text.write(line[j][0])
                
                text.write
                
        text.write('\n')
                
            

```
